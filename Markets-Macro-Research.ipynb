```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
inflation = pd.read_csv('/Users/rahulyadav/Downloads/inflation_data.csv')
bond_yields = pd.read_csv('/Users/rahulyadav/Downloads/bond_yields_data.csv')

# Merge datasets
data = pd.merge(inflation, bond_yields, on=['Country', 'Year'])

# Quick Look
print(data.head())

# Plot Inflation Trend
plt.figure(figsize=(10,5))
sns.lineplot(data=data, x='Year', y='Inflation', hue='Country', marker='o')
plt.title('Inflation Trend (2019–2024)')
plt.ylabel('Inflation Rate (%)')
plt.xlabel('Year')
plt.grid(True)
plt.show()

# Plot Bond Yields Trend
plt.figure(figsize=(10,5))
sns.lineplot(data=data, x='Year', y='Bond_Yield', hue='Country', marker='o')
plt.title('Bond Yields Trend (2019–2024)')
plt.ylabel('Bond Yield (%)')
plt.xlabel('Year')
plt.grid(True)
plt.show()

# Correlation between Inflation and Bond Yields by Country
for country in data['Country'].unique():
    country_data = data[data['Country'] == country]
    corr = country_data['Inflation'].corr(country_data['Bond_Yield'])
    print(f"Correlation between Inflation and Bond Yield in {country}: {corr:.2f}")
```
